```python
dump = pl.read_parquet('data/online_diss_cleaned_03-2024.parquet') 
data = dump.to_pandas()
#data = pd.read_csv('data/check-data-csv.csv')
#data
```

```python
# Copy DDC3 to DDC where DDC = N/A: 
data_ddc = data.rename(columns={"DDC-DE?":"DDC3"})
data_ddc["DDC"] = np.where((data_ddc.DDC == "N/A"), data_ddc.DDC3, data_ddc.DDC)
data_ddc["DDC"] = np.where((data_ddc.DDC == "N/A"), data_ddc.Sachgruppe, data_ddc.DDC)
data_ddc
```

```python
mainddc = merged['DDC'].str[:3]  # create list of 1. ddc entry (and shorten DDC to first 3)
data2 = pd.DataFrame(mainddc)  # convert to dataframe
data2["count"] = data2.groupby("DDC")["DDC"].transform('count') # add column for value_count of DDC-number
data2
```

```python
data2 = data2.drop_duplicates("DDC") # drop duplicate entries based on DDC-number
data2 = data2.rename(columns={"DDC": "DDCthird"})
print(len(data2))
```
