```python
import pandas as pd
```

```python
data = pd.read_hdf('data/online-diss-cleaned_11-2023.h5', 'df')
#data = pd.read_csv('data/check-data-csv.csv')
data
```

```python
stats = data[["ID", "lang", "Year"]]
#stats[stats['Year'].isnull()]
#stats
```

```python
lang = stats.groupby(['lang']).size().reset_index(name='counts')
sprachen = pd.read_excel("data/Sprachen.xlsx")
sprachen
```

```python
lang = pd.merge(lang, sprachen, on=['lang'], how="left")
lang
```

```python
lang = lang[lang.lang != "und"]
lang = lang[lang.lang != "fail"]
lang
```

```python
lang.to_json("data/diss_lang.json", index=True)
```

```python
all_lang = (lang['counts']).sum()
print(all_lang)
```
